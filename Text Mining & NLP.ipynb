```python
M=nltk.corpus.movie_reviews.raw('neg/cv985_5964.txt')
M
```

```python
# tokenization
```

```python
from nltk.tokenize import word_tokenize, sent_tokenize
```

```python
tokens=word_tokenize(M)
tokens
```

```python
from nltk.probability import FreqDist
```

```python
fdist=FreqDist()
for i in tokens:
    fdist[i.lower()]+=1
fdist    
```

```python
fdist_top10=fdist.most_common(10)
fdist_top10
```

```python
from nltk.tokenize import blankline_tokenize
```

```python
# punctuation


import re
punctuation=re.compile(r'[.,-,?!,:;()|0-9]')

post_punctuation=[]
for words in tokens:
    word=punctuation.sub("",words)
    if len(word)>0:
        post_punctuation.append(word)
```

```python
from nltk import ne_chunk
```

```python
ne_tags=nltk.pos_tag(tokens)
ne_tags
```

```python
# chunking
```

```python
ne_NameEntityRecognition=ne_chunk(ne_tags)
ne_NameEntityRecognition
```

```python
grammar_np=r'NP : {<DT>?<JJ>*<NN>}'
chunk_parser=nltk.RegexpParser(grammar_np)
```

```python
chunk_result=chunk_parser.parse(ne_tags)
chunk_result
```
